```python
import pandas as pd
import numpy as np
```

# İleri seviye Birleştirme işlemleri

```python
df1 = pd.DataFrame({"Calisanlar": ["Ali", "Veli", "Ayşe", "Fatma"],
                   "grup": ["Muhasebe","Muhendislik","Muhendislik","I.Kaynaklari"]})
df1
```

```python
df2 = pd.DataFrame({"Calisanlar": ["Ali", "Veli", "Ayşe", "Fatma"],
                   "ilk_giris": [2010,2009,2016,2019]})
df2
```

```python
df3 = pd.merge(df1,df2)
# Normalde neye göre birleştireceğini kendisi anlıyor.
df3
```

```python
# İstersek kendimiz belirleyebiliriz.
pd.merge(df1, df2, on="Calisanlar")
```

# Birleştirmelerde gruplandırma
## Many to one Merge (Çoktan Teke Birleştirme)

```python
df3
```

```python
df4 = pd.DataFrame({"grup": ["Muhasebe", "Muhendislik","I.Kaynaklari"],
                   "mudur": ["Caner", "Mustafa", "Berkcan"]})
df4
```

```python
pd.merge(df3, df4)
```

## Many to Many Merge (Çoktan Çoka Birleştirme)

```python
df5 = pd.DataFrame({"grup": ["Muhasebe", "Muhasebe",
                            "Muhendislik","Muhendislik",
                            "I.Kaynaklari","I.Kaynaklari"],
                   "Yetenekler": ["matematik", "excel", "kodlama",
                                 "linux", "excel", "yonetim"]})
df5
```

```python
df1
```

```python
pd.merge(df1, df5)
```
