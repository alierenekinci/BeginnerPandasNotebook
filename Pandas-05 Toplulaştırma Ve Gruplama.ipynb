# Gruplama İşlemleri

```python
import pandas as pd 
```

```python
df = pd.DataFrame({
    "gruplar" : ["A", "B", "C", "A", "B","C"],
    "veri" : [10,11,52,23,43,55]
}, columns=["gruplar", "veri"])
```

```python
df
```

```python
df.groupby("gruplar")
```

```python
# Toplulaştırma fonksiyonu istiyor.
df.groupby("gruplar").mean()
```

```python
# Toplulaştırma fonksiyonu istiyor.
df.groupby("gruplar").count()
```

```python
# Toplulaştırma fonksiyonu istiyor.
df.groupby("gruplar").std()
```

Çok iyi değilmi :)

```python
df.groupby("gruplar").describe().T
```

# İleri Toplulaştırma İşlemler (Aggregate, Filter, transform, apply)

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "gruplar": ["A","B","C","A","B","C"],
    "degisken1": [10,23,33,22,11,99],
    "degisken2": [100,253,333,262,111,969]
}, columns=["gruplar", "degisken1","degisken2"])
df
```

## aggregate

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate(["min", "median", "max"])
```

```python
df.groupby("gruplar").aggregate({"degisken1": ["min", "max"]})
```

## Filter

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "gruplar": ["A","B","C","A","B","C"],
    "degisken1": [10,23,33,22,11,99],
    "degisken2": [100,253,333,262,111,969]
}, columns=["gruplar", "degisken1","degisken2"])
df
```

```python
def randomFunc(x):
    return x["degisken1"].mean() <= 33
```

```python
df.groupby("gruplar").filter(randomFunc)
```

## transform

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "gruplar": ["A","B","C","A","B","C"],
    "degisken1": [10,23,33,22,11,99],
    "degisken2": [100,253,333,262,111,969]
}, columns=["gruplar", "degisken1","degisken2"])
df
```

```python
df.degisken1*9
```

```python
df_a = df.iloc[:,1:]
df_a
```

```python
df_a.transform(lambda x: (x-x.mean()) / x.std())
```

## Apply

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "gruplar": ["A","B","C","A","B","C"],
    "degisken1": [10,23,33,22,11,99],
    "degisken2": [100,253,333,262,111,969]
}, columns=["gruplar", "degisken1","degisken2"])
df
```

```python
df.apply(np.sum)
```

```python
df.groupby("gruplar").apply(np.mean)
```
